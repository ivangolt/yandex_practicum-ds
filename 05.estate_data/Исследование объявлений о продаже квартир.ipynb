```python
good_data['price_per_sq_m'] = (good_data['last_price'] / good_data['total_area']).round().astype('int')

column_name = 'price_per_sq_m'
column_values = good_data.pop(column_name)  # Удаляем столбец из конца и сохраняем его значения
good_data.insert(3, column_name, column_values) 

good_data.head()
```

```python
day_of_week_mapping = {
    0: 'Понедельник',
    1: 'Вторник',
    2: 'Среда',
    3: 'Четверг',
    4: 'Пятница',
    5: 'Суббота',
    6: 'Воскресенье'
}

good_data['day_of_week'] = good_data['first_day_exposition'].dt.dayofweek.map(day_of_week_mapping)

column_name = 'day_of_week'
column_values = good_data.pop(column_name)
good_data.insert(5,column_name, column_values)
good_data.head()
```

```python
good_data.loc[:, 'first_day_exposition'].dt.month
```

```python
month_names = {
    1: 'Январь',
    2: 'Февраль',
    3: 'Март',
    4: 'Апрель',
    5: 'Май',
    6: 'Июнь',
    7: 'Июль',
    8: 'Август',
    9: 'Сентябрь',
    10: 'Октябрь',
    11: 'Ноябрь',
    12: 'Декабрь'
}

good_data['month'] = good_data['first_day_exposition'].dt.month.map(month_names)

column_name = 'month'
column_values = good_data.pop(column_name)
good_data.insert(6,column_name, column_values)
good_data.head()
```

```python
good_data['year'] = good_data['first_day_exposition'].dt.year

column_name = 'year'
column_values = good_data.pop(column_name)
good_data.insert(7,column_name, column_values)
good_data.head()
```

```python
def categorize_floor(row):

    floor = row['floor']
    floor_total = row['floors_total']
    
    if floor == 1:
        return 'первый'
    
    if floor == floor_total:
        return 'последний'
    
    return 'другой'

good_data['floor_type'] = good_data.apply(categorize_floor,axis=1)
column_name='floor_type'
column_values = good_data.pop(column_name)
good_data.insert(13,column_name, column_values)
```

```python
good_data['last_price_millions'] = good_data['last_price'] / 1000000
```

```python
good_data_last_price = good_data[['last_price','last_price_millions','total_area','living_area','kitchen_area','rooms','floor_type','day_of_week','month','year']]
good_data_last_price.head()
```

Как можно заметить, цена объекта положительно коррелирует с общей площадью объекта, жилой площадью, площадью кухни и количеством комнат. И никак не коррелирует с годом размещения обьъекта.

```python
columns_to_analyze = ['total_area', 'living_area', 'kitchen_area']

fig, axes = plt.subplots(nrows=1, ncols=len(columns_to_analyze), figsize=(15, 5))

for i, column in enumerate(columns_to_analyze):
    x = good_data_last_price[column]
    y = good_data_last_price['last_price']
    
    axes[i].scatter(x, y, alpha=0.5)
    axes[i].set_title(f'Зависимость цены от {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Цена')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

Коэффициент корреляции 0.728205 означает, что существует довольно сильная положительная линейная связь между общей площадью квартиры и её ценой. То есть, как правило, с увеличением общей площади квартиры цена также увеличивается. Это может быть объяснено тем, что квартиры с большей площадью обычно имеют больше места и удобств, что может повысить их стоимость.\
