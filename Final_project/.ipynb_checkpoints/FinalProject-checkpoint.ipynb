<span style="color:black;font-weight:700;font-size:22px"> 
              Описание проекта
</span>

Чтобы оптимизировать производственные расходы, металлургический комбинат «Стальная птица» решил уменьшить потребление электроэнергии на этапе обработки стали. Для этого комбинату нужно контролировать температуру сплава. Ваша задача — построить модель, которая будет её предсказывать. 
Заказчик хочет использовать разработанную модель для имитации технологического процесса.

**Описание этапа обработки**\
Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша.\ 
Из сплава выводится сера (этот процесс — десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (от англ. tribe — «масса»).\
Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.\
Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (от англ. slab — «плита»).

## data_arc_new

```python
data_arc_new_df.head()
```

В данном датасете вычислим суммарную мощность по формуле: 
$$ Полнаямощность = \sqrt ((Активная мощность)^2 + (Реактивная мощность)^2) $$
    Также после этого агрегируем суммарную мощность по всей партии


Данные о времени не представляют интереса для анализа и будут удалены.

```python
data_arc_new_df = data_arc_new_df.rename(columns={'Начало нагрева дугой': 'heat_start_time', 
                                                  'Конец нагрева дугой':'heat_end_time',
                                                  'Активная мощность': 'active_power',
                                                  'Реактивная мощность': 'reactive_power'})
```

```python
data_arc_new_df.head()
```

```python
data_arc_new_df[['active_power', 'reactive_power']].hist(bins=50, figsize=(10,3));
```

Перед агрегированием удалим аномалии в виде отрицательных значений мощности.

```python
data_arc_new_df = data_arc_new_df[(data_arc_new_df['reactive_power'] >=0) & (data_arc_new_df['active_power'] >= 0)]
```

```python
data_arc_new_df[['active_power', 'reactive_power']].hist(bins=50,figsize=(10,3));
```

```python
data_arc_new_df['full_power'] = (((data_arc_new_df['active_power']) ** 2) + ((data_arc_new_df['reactive_power'])**2)) ** 0.5
```

```python
data_arc_new_df.head(5)
```

Также добавим признак работы выполненой электродами за время нагревания.
Вычислим ее по формуле **A = N*t** , где N - полная мощность, время в секундах и время нагрева дуги

```python
data_arc_new_df['heat_time,s'] = (data_arc_new_df['heat_end_time'] - data_arc_new_df['heat_start_time']).dt.total_seconds()
```

```python
data_arc_new_df.head()
```

Теперь агрегируем полную мощность за всю партию

```python
data_arc_new_df_agr = data_arc_new_df.pivot_table(index='key',values=['full_power', 'heat_time,s'], aggfunc=np.sum)
```

```python
data_arc_new_df_agr.head()
```

```python
data_arc_new_df_agr['perfomance'] = data_arc_new_df_agr['full_power'] * data_arc_new_df_agr['heat_time,s']
```

Также добавим столбец, который будет показывать количество запусков нагрева 

```python
count_electrod = data_arc_new_df.pivot_table(index='key', values='active_power', aggfunc='count')\
                                .rename(columns={"active_power": 'heating_quantity'})
count_electrod
```

```python
data_arc_new_df_agr['heating_quanity'] = count_electrod['heating_quantity']
```

```python
data_arc_new_df_agr.head()
```

```python
plot_violin(data_arc_new_df_agr, 'full_power')
```

Для данного датасета были вычислены полная мощность в дискретный отсчет времени в течении партии. После чего была вычеслена суммарная партия за всю партию. Также в этом датасете былы удалены аномалии в виде отрицательных значений мощностей.

Обработаем 0 пропуски в данном датасете и с просуммируем общий объем сыпучих материалов за партию

```python
def snake_case_column_names(df):

    def to_snake_case(s):
        return s.strip().lower().replace(' ', '_')

    df.columns = [to_snake_case(col) for col in df.columns]

    return df
```

# Подготовка данных

На данном этапе подготовим данные для обучения моделей машинного обучения. Для этого разделим выборку на тренировочную и тестовые выборки, выделим target переменную, а также подготовим пайплайн предобработки данных.

```python
RANDOM_STATE = 40324
TEST_SIZE = 0.25
```

```python
y = df['temperature']
X = df.drop(columns=['temperature'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE)
print(f'X_train.shape: {X_train.shape}, y_train_shape: {y_train.shape}')
print(f'X_test.shape: {X_test.shape}, y_train_shape: {y_test.shape}')
```

```python
num_cols = X.columns
num_col_pipeline = Pipeline(steps=[
    ('num_scaler', StandardScaler()),
#     ('polynomial', PolynomialFeatures(degree=3))
])

data_preprocessor = ColumnTransformer([('num', num_col_pipeline, num_cols)],
                                      remainder='passthrough')

model_pipeline = Pipeline(steps=[
    ('data_preprocessing', data_preprocessor),
    ('regression_model', LinearRegression())
])

param_grid = [
    {'regression_model': [LinearRegression()]},
    
    {'regression_model': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
     'regression_model__max_depth': [5, 10, 15],
     'regression_model__min_samples_leaf': [5, 10, 15],
     'regression_model__min_samples_split': [5, 10, 15]
    },
    
    {'regression_model':[KNeighborsRegressor()],
     'regression_model__n_neighbors': range(5,20)
    },
    
    {'regression_model': [RandomForestRegressor(random_state=RANDOM_STATE)],
     'regression_model__n_estimators': [100, 200, 500]
    },
    
    {'regression_model': [CatBoostRegressor(random_state=RANDOM_STATE, verbose=False)],
     'regression_model__learning_rate': [0.001, 0.01, 0.1, 1],
     'regression_model__iterations': [100, 200, 400, 800]
    },
    
    {'regression_model': [LGBMRegressor(random_state=RANDOM_STATE, verbose=-100)],
     'regression_model__n_estimators': [100, 200, 400, 800, 1000],
     'regression_model__learning_rate': [0.001, 0.01, 0.1, 1]
    },
    
    {'regression_model': [XGBRegressor(random_state=RANDOM_STATE)],
     'regression_model__n_estimators': [100, 200, 400, 800, 1000],
     'regression_model__learning_rate': [0.001, 0.01, 0.1, 1]}
]
```

<span style="color:black;font-weight:700;font-size:17px"> 
              Вывод
</span>

На данном этапе был подготовлен пайплайн обработки данных, который включает в себя этапы масштабирования признаков и перебор гиперпарметров моделей.

# Обучение моделей

```python

grid_search = GridSearchCV(model_pipeline,
                          param_grid,
                          n_jobs=-1,
                          cv=10,
                          scoring='neg_mean_absolute_error')

grid_search.fit(X_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
print(f'Лучшая модель: {grid_search.best_estimator_}')
print(f'Лучшая метрика на кросс-валидации:{grid_search.best_score_}')
```

<span style="color:black;font-weight:700;font-size:17px"> 
              Вывод
</span>

На данном этапе были обучены модели с подбором гиперпараметров по сетке gridsearch. Для обучения были выбраны следующие модели: 
- Линейная регрессия
- Дерево решений
- метод k-ближайших соседей
- Cлучайный лес
- Catboost
- LightGBM
- XGBoost

Наилучший результат на кросс-валидации показала модель LGBMRegressor(learning_rate=0.01, n_estimators=400,
                               random_state=40324, verbose=-100) 
Метрика модели на кросс-валидации MAE = 6.33
                               

```python
best_model = grid_search.best_estimator_
```

# Выбор лучшей модели

На данном этапе выберем лучшую модель и предскажем тестовые значения для тестовой выборки.А также проведем анализ значимости факторов т.е. определим какие признаки вносят наибольший вклад в результаты модели.

```python
y_predict = best_model.predict(X_test)
```

```python
print(f'MAE на тестовой выюорке: {mean_absolute_error(y_predict, y_test)}')
```

```python
def get_quality_and_metrics_of_model(predictions, y_test):
    mae = mean_absolute_error(predictions, y_test)
    mse = mean_squared_error(predictions, y_test)
    rmse = mean_squared_error(predictions, y_test, squared=False)
    print(f'MAE = {mae:.0f}')
    print(f'MSE = {mse:.0f}')
    print(f'RMSE = {rmse:.0f}')
    
    
    residuals = y_test - predictions
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,10))
    axes[0].hist(residuals, bins=30)
    axes[0].axvline(x=0, color='red', linestyle='--', label='Линия нуля')
    axes[0].axvline(x=residuals.mean(), color='yellow', linestyle='dotted', label='Линия среднего')
    axes[0].text(0, 10, f'Mean Residuals = {residuals.mean():.2f}', color='black', fontsize=8)
    axes[0].set_title('Гистограмма распределения остатков')
    axes[0].legend()
    axes[0].set_xlabel('Остатки')

    axes[1].scatter(predictions, residuals)
    axes[1].set_xlabel('Предсказания модели')
    axes[1].set_ylabel('Остатки')
    axes[1].set_title('Анализ дисперсии')
    plt.show()
```

```python
get_quality_and_metrics_of_model(y_predict, y_test)
```

Анализ остатков показывает, что среднее значение смещено влево от нуля, и говорит о том, что модель дает большие предсказания , её прогнозы больше истинных значений
